```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("./medical_insurance_cost_dataset.csv")
```

```python
df.head(10)
```

```python
# Separate Independent (X) and Dependent (Y) features
X = df.drop(columns=['annual_medical_cost_usd'])
Y = df['annual_medical_cost_usd']

# Print shapes
print("Shape of X (Independent Features):", X.shape)
print("Shape of Y (Dependent Feature):", Y.shape)
```

```python
X
```

```python
print("\nBefore One-Hot Encoding")
print(df[['gender','smoker','region','children']])
df_encoded=pd.get_dummies(df[['gender','smoker','region','children']])
df_encoded=df_encoded.astype(int)
print("\nAfter one hot encoding")
print(df_encoded.head(5))
```

```python
from sklearn.preprocessing import StandardScaler
numeric_feature=['age','bmi','annual_medical_cost_usd']
scaler=StandardScaler()
df[numeric_feature]=scaler.fit_transform(df[numeric_feature])
print("\nFirst 5 rows after standard scaling:")
print()
print(df[numeric_feature].head(5))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.32, random_state=42)
print("Shape of X_train:", X_train.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of Y_train:", Y_train.shape)
print("Shape of Y_test:", Y_test.shape)
```

```python
X = pd.get_dummies(df.drop("annual_medical_cost_usd", axis=1), drop_first=True)
Y = df["annual_medical_cost_usd"]
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
y_pred=model.predict(X_test)
print("Intercept")
print(model.intercept_)
print()
print("Coefficients")
print(model.coef_)
```

```python
y_pred
```

```python
from sklearn.metrics import  r2_score

# Make predictions on the test set
print("\nPredicted Value:",y_pred[0: 11])
print("\nActual Value:",Y_test.tolist()[0: 11])
```

```python
from sklearn.metrics import  r2_score
r2=r2_score(Y_test,y_pred)
print("R2 Score:",r2)
```

```python
# The model achieved an R² score of 0.888,
# indicating that approximately 89% of the variation in medical insurance costs is explained by the selected features.
#  This suggests the model is performing very well, though some variability remains unexplained.”
```
